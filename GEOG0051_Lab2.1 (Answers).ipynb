### Basic Operations

### Broadcasting

In NumPy you can perform operations on arrays of different shapes through broadcasting. Broadcasting provides a means of vectorizing array operations. The lower dimension array gets stretched to the higher dimension one to perform an operation. For example, the outputs in this cell is equivalent.

```python
np.array([1, 2]) * 2
```

Please read the [following](https://numpy.org/doc/stable/user/basics.broadcasting.html) for further information.

````python
# basic operations on arrays. 
array0 = np.array([1, 2, 3])
print (f'array0 + 1 : {array0 + 1}')
print (f'array0 * 2 : {array0 * 2}')
print (f'array0 **2 : {array0 ** 2}')
print (f'log(array) : {np.log(array0)}')
````

### Merging Data Frames

There are multiple ways to join two pandas data frames in Python, and these are summarised in the chart below. The merging of data frames usually occurs based on a column/variable that is common to both datasets. The main difference between the kinds of merges is **which combination rows get kept** in the output data frame, and these are elaborated on below.

```python
how : {'left', 'right', 'outer', 'inner'} # default type of merge to be performed = 'inner' 
```

* **Left:** Uses only keys (i.e. keeps only the values) from left data frame, similar to a SQL left outer join;
  preserve key order.
* **Right:** Uses only keys from right data frame, similar to a SQL right outer join;
  preserve key order.
* **Outer:** Uses union of keys (i.e. keeps all values) from both data frames, similar to a SQL full outer
  join; sort keys lexicographically.
* **Inner:** Uses intersection of keys from (i.e. keeps only values common to) both data frames, similar to a SQL inner
  join; preserve the order of the left keys.
  
      
    


````python
dflat=df[['longitude']] #subsets specific columns of the original data frame
dflon=df[['latitude']]
````

````python
### you can also join/merge them together based on a common index
dfBoth = pd.merge(left=dflon, right=dflat, left_index=True, right_index=True)
````

### Subplots

Sometimes it is helpful to have different plots of the same or related datasets side-by-side. For this purpose, Matplotlib allows us the possibility of plotting 'subplots' within a larger plot: individual plots that can exist together within a single figure. 


````python
fig, ax = plt.subplots(figsize=(10,6)) 
for i in range(1, 7):
    #plt.subplot(nrows, ncolumns, current subplot)
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((i)),
             fontsize=18, ha='center')
fig.tight_layout(pad=2.0)
plt.show()
````

### Visualising data directly from a Pandas data frame

Rather than invoking the `hist()` function from the `PyPlot` sub-package within MatPlotLib, a functionality that Pandas data frames and series afford is that of a method (built-in function) to plot charts of different sorts **directly from the data frame/series itself!**. To do this, we use the general structure:

```python

```

And have a look at the [function documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html) for information of what `kind=`s of plots you can make.

````python
#https://gist.github.com/perrygeo/c426355e40037c452434
from scipy import ndimage
def heatmap(d, bins=(100,100), smoothing=1.3, cmap='jet', alpha=0.6):
    def getx(pt):
        return pt.coords[0][0]

    def gety(pt):
        return pt.coords[0][1]

    x = list(d.geometry.apply(getx))
    y = list(d.geometry.apply(gety))
    heatmap, xedges, yedges = np.histogram2d(y, x, bins=bins)
    extent = [yedges[0], yedges[-1], xedges[-1], xedges[0]]

    logheatmap = np.log(heatmap)
    logheatmap[np.isneginf(logheatmap)] = 0
    logheatmap = ndimage.filters.gaussian_filter(logheatmap, smoothing, mode='nearest')
    
    plt.imshow(logheatmap, cmap=cmap, extent=extent, alpha=0.6)
    plt.colorbar()
    plt.gca().invert_yaxis()
    plt.show()
     
    return logheatmap,extent
````

````python
ax = gdf.plot(figsize=(20,20),color='white')
logheatmap,extent=heatmap(gdf, bins=50, smoothing=1.5, alpha=0.6)
````
